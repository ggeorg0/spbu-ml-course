Переведем некоторые **категориальные** переменные в **количественные** с помощью метода `pandas.Series.map()`

```python
mapping_cols = (
    'Parental_Involvement',
    'Access_to_Resources',
    'Motivation_Level',
    'Family_Income',
    'Teacher_Quality',
)

keys = ('Low', 'Medium', 'High')
map_rules = dict(zip(keys, range(len(keys)))) # Low: 0, 'Medium': 1... 

for col in mapping_cols:
    # I am using `df.loc[:, col] =` here instead of `df[col] =`
    #  because "Copy-on-Write" will become the new default in pandas 3.0.
    #  and `df[col]` produces a warning.
    # df.loc[:, col] = df[col].map(map_rules).astype('int64')
    # print(df[col].map(map_rules).head())
    df[col] = df[col].map(map_rules).astype('int8')

df.head(5)
```

```python
df.dtypes
```

Можно задать отдельную функцию которая принимает на вход названия колонок и правила, по которым мы "сопоставляем" значения.

С помощью новой фукнции переведем все колонки бинарными значениями "Yes"/"No" к числовым `1` / `0`

```python
def map_cols(df, columns, rules, astype='int8'):
    """Map values in given columns with `rules` dict.
    This function also changes the given DataFrame `df` 
    and returns nothing.
    """
    for col in columns:
        df[col] = df[col].map(rules).astype(astype)


binary_cols = (
    'Internet_Access',
    'Learning_Disabilities',
    'Extracurricular_Activities'
)
map_cols(df, binary_cols, {'Yes': 1, 'No': 0})

df.head(5)
```

Для оставшихся колонок выведем их уникальные значения.

```python
remaning_cols = (
    'School_Type',
    'Peer_Influence',
    'Parental_Education_Level',
    'Distance_from_Home',
    'Gender'
)

print('-' * 70)
for col in remaning_cols:
    print(f"{col:30s} | {df[col].unique()}")
```

Теперь мы знаем, какие значения находятся в оставшихся колонках. Их можно тоже перевести в числовой вид. 

(Вообще говоря это делать необязательно, потому что теряется смысловая связь с исходными данными, но мне хочется чтобы везде были числа!)

```python
from functools import partial

map_df_cols = partial(map_cols, df)

map_df_cols(('School_Type', ), {'Public': 1, 'Private': 0})
map_df_cols(('Peer_Influence', ), {'Positive': 1, 'Negative': -1, 'Neutral': 0})
map_df_cols(('Parental_Education_Level', ), {'High School': 0, 'College': 1, 'Postgraduate': 2})
map_df_cols(('Distance_from_Home', ), {'Near': 0, 'Moderate': 1, 'Far': 2})
map_df_cols(('Gender', ), {'Male': 0, 'Female': 1})

df
```

### Feature engeneering

Презжде чем перейти к фьюче инженерингу, посмотрим ещё на признак `"Previous_Scores"` и оценим его коэффициент корреляции

```python
df["Previous_Scores"]
```

```python
df[["Previous_Scores", "Exam_Score"]].corr()
```

Коэффициент корреляции не очень большой, однако положительный. 

Есть идея **соеденить два признака**, например `"Previous_Scores"` и `"Hours_Studied"` — кажется, что если у ученика было много баллов на прошлом экзамене и он тратит много времени на подготовку, то и на предстоящем экзамене он тоже получит хорошую оценку.
Назовем новый признак `"Prep_n_LastExam"`

Во-первых, попробуем просто сложить два признака и посмотрим на коэффициент корреляции.

```python
df["Prep_n_LastExam"] = df["Previous_Scores"] + df["Hours_Studied"]
df["Prep_n_LastExam"]
```

```python
df[["Prep_n_LastExam", "Exam_Score"]].corr()
```

К сожалению, коэффициент корреляции получается даже меньше, чем просто для признака `"Hours_Studied"`. 
Так происходит потому, что признак `"Previous_Scores"` вносит слишком много шума, ведь коэффиценты корреляций складываемых признаков неодинаковы.

Попробуем перемножить значения признаков.

```python
df["Prep_n_LastExam"] = df["Previous_Scores"] * df["Hours_Studied"]

corr_before = df[["Hours_Studied", "Exam_Score"]].corr().iloc[0, 1]
corr_feature_eng = df[["Prep_n_LastExam", "Exam_Score"]].corr().iloc[0, 1]

print("Hours_Studied corr:", corr_before)
print("Prep_n_LastExam corr:", corr_feature_eng)
```

Получился коэффициент корреляции больше, чем у `Hours_Studied` изначально.\
Однако прирост составил всего на `~0.02`, что не очень много. 

Попробуем объеденить другие два признака, которые давали самую большую корреляцию: `Hours_Studied` и `Attendance`.

Новый признак назовем `Engagement` — что значит "вовлеченность"

```python
df["Engagement_mul"] = df["Hours_Studied"] * df["Attendance"]
df["Engagement_sum"] = df["Hours_Studied"] + df["Attendance"]

corr_mul = df[["Engagement_mul", "Exam_Score"]].corr().iloc[0, 1]
corr_sum = df[["Engagement_sum", "Exam_Score"]].corr().iloc[0, 1]

print("Corr. `Hours_Studied` * `Attendance` =", corr_mul)
print("Corr. `Hours_Studied` + `Attendance` =", corr_sum)
```

**Ого!** Новый при сложении `Hours_Studied` с `Attendance` мы получаем новый признак с корреляцией 0.72! 
Это значительно больше, чем все признаки которые у нас были до этого:

```python
df.corr()['Exam_Score']
```

Посмотрим на точечную диаграмму для средней оценки по новому признаку 

```python
mean_exam_score = df.groupby(by="Engagement_sum")["Exam_Score"].agg(["mean"])

plt.scatter(x=mean_exam_score.index, y=mean_exam_score, color='red')
plt.xlabel("Engagement_sum")
plt.ylabel("Mean exam score")
plt.show()
```

Таким образом новый признак `Engagement_sum` можно использовать для обучения модели
